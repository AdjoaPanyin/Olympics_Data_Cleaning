### Handle null of age, height and weight

```python
#Calculate the median age
#the reason why we are not using mean but median is because the mean is sensitive to outliers
median_age= df_new["Age"].median()

#fill NaN values with the median age
df_new["Age"]= df_new["Age"].fillna(median_age)
```

```python
df_new["Age"].unique()
```

```python
#Calculate the mean height
mean_height= df_new["Height"].mean()

#fill NaN values with the mean height
df_new["Height"]= df_new["Height"].fillna(mean_height)
```

```python
df_new["Height"].unique()
```

```python
#Calculate the mean weight
mean_weight= df_new["Weight"].mean()

#fill NaN values with the mean weight
df_new["Weight"]= df_new["Weight"].fillna(mean_weight)
```

```python
df_new["Weight"].unique()
```

```python
df_new.info()
```

```python
#converting the datatypes
df_new['Sex']=df_new['Sex'].astype('category')
df_new['Age']=df_new['Age'].astype('int')
df_new['Team']=df_new['Team'].astype('category')
df_new['Season']=df_new['Season'].astype('category')
df_new['Sport']=df_new['Sport'].astype('category')
df_new['NOC']=df_new['NOC'].astype('category')
df_new['Medal']=df_new['Medal'].astype('category')
```

```python
#converting the datatypes
df_new['male']=df_new['male'].astype('category')
df_new['education']=df_new['education'].astype('category')
df_new['currentSmoker']=df_new['currentSmoker'].astype('category')
df_new['BPMeds']=df_new['BPMeds'].astype('category')
df_new['prevalentStroke']=df_new['prevalentStroke'].astype('category')
df_new['prevalentHyp']=df_new['prevalentHyp'].astype('category')
df_new['diabetes']=df_new['diabetes'].astype('category')
df_new['TenYearCHD']=df_new['TenYearCHD'].astype('category')
```

#### Feature Engineering

```python
df_height_in_meters= df_new["Height"] / 100
df_new["BMI"]= df_new["Weight"]/df_height_in_meters **2
```

```python
df_new[["BMI"]]
```

```python
#Round numbers in the 'BMI' column to the nearest whole number
df_new['BMI_rounded']= df_new['BMI'].apply(round)
```

```python
#ANALYSING THE RELATIONSHIP BETWEEN BMI AND MEDAL
#groupby function
df_new.groupby(df_new["BMI_rounded"])["Medal"].value_counts().to_frame()
```

```python
np.random.seed(0)
data = {
    'Medal': np.random.choice(['Bronze', 'Silver', 'Gold','no_medal'], size=1000),
    'BMI': np.random.normal(loc=25, scale=5, size=1000)}  # Example BMI values

# Convert data to a DataFrame
df = pd.DataFrame(data)

# Plotting with Seaborn - Box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Medal', y='BMI', data=df, palette='Set2')
plt.title('BMI Distribution by Medal Type')
plt.xlabel('Medal Type')
plt.ylabel('BMI')

# Show plot
plt.show()
```
